Example of how to use cooling function table of <a href="http://adsabs.harvard.edu/abs/2012ApJS..202...13G">Gnedin & Hollon (2012)</a>. This function is shown below by the red line for the solar composition and can be compared with the cooling curve for primordial composition plasma (blue curve). The table <a href="http://astro.uchicago.edu/~andrey/prj/artdisco/test_n10.0.res"><tt>test_n10.0.res</tt></a> containts the tabulated function of Gnedin & Hollon 2012 for the primordial composition ($Z=0$) and $Z=Z_\odot$. The table can be interpolated for intermediate temperatures. The table was produced with fortran codes described <a href="http://astro.uchicago.edu/~gnedin/CF/chf.html">here</a> for gas density of $n=10\, \mathrm{cm^{-3}}$. They can be used to produce similar tables for other densities and metallicities.

```python
import numpy as np
 
# setup notebook for inline figures
%matplotlib inline

# import pyplot and set some parameters to make plots prettier
import matplotlib.pyplot as plt

plt.rc("savefig",dpi=175)
plt.rc('text', usetex=True)
plt.rc('font',size=9)
plt.rc('xtick.major',pad=5); plt.rc('xtick.minor',pad=5)
plt.rc('ytick.major',pad=5); plt.rc('ytick.minor',pad=5)
lTgh, cf0gh, hf0gh, cf1gh, hf1gh = np.loadtxt(r'test_n10.0.res',unpack=True) 
#
# plot
#
fig = plt.figure(figsize=(4, 4))
 
plt.xlabel(r'$T (K)$')
plt.ylabel(r'$\Lambda(T)\ \rm (ergs\ s^{-1}\ cm^3)$')
plt.xscale('log'); 
plt.yscale('log')
plt.xlim(100., 1.e8)
plt.ylim(1.e-28, 1.e-20)
 
plt.plot(10.**lTgh, cf0gh-hf0gh, c='b', lw=2.0, label=r'Gnedin \& Hollon 2012 $(Z=0)$')
plt.plot(10.**lTgh, cf1gh-hf1gh, c='r', lw=2.0, label=r'Gnedin \& Hollon 2012 $(Z=Z_\odot)$')
 
plt.legend( frameon=False, loc='upper left',fontsize=9)
 
plt.grid()
plt.show()
 
```
